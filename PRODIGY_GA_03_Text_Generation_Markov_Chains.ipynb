```python
text_data = """
Artificial intelligence is transforming the world.
Machine learning enables systems to learn from data.
Data science combines statistics and programming.
Artificial intelligence and machine learning are closely related.
"""
```

```python
# Convert text to lowercase and split into words
words = text_data.lower().split()

print(words)
```

```python
# Create Markov chain dictionary
markov_chain = {}

for i in range(len(words) - 1):
    current_word = words[i]
    next_word = words[i + 1]

    if current_word not in markov_chain:
        markov_chain[current_word] = []

    markov_chain[current_word].append(next_word)

# Print the Markov chain
markov_chain
```

```python
import random

def generate_text(start_word, length=20):
    current_word = start_word.lower()
    result = [current_word]

    for _ in range(length - 1):
        if current_word not in markov_chain:
            break

        next_word = random.choice(markov_chain[current_word])
        result.append(next_word)
        current_word = next_word

    return " ".join(result)
```

```python
print(generate_text("artificial", 20))
```

```python
print(generate_text("machine", 20))
print(generate_text("data", 20))
```
